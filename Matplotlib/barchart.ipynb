```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv("sales_data_sample.csv",encoding="Latin")
data = data.drop(columns=['ORDERNUMBER', 'ORDERDATE', 'PRODUCTCODE', 
    'CUSTOMERNAME', 'PHONE', 'ADDRESSLINE1', 'ADDRESSLINE2',
    'CONTACTLASTNAME', 'CONTACTFIRSTNAME'])
print(data.head())
```

```python
print(data['PRODUCTLINE'].unique()) # this gives the unique values in the col 
cat1 =data['PRODUCTLINE'].value_counts() #we get the total count of all unique values in this catagory in assending order 
print(cat1)
# cat1.value_counts()
```

## There for by .value_count 
### we get the top 4 ProductLIne these are classic cars , vintage cars , motorcycles or planes 

```python
x_bar = cat1.index
print(x_bar)
y_bar = cat1.values
result=plt.bar(x_bar,y_bar)
print(result)
```

```python
plt.figure(figsize=(12,8))
plt.title("Product Line Distribution", fontsize=14)
plt.xlabel("Product Line")
plt.ylabel("count")
plt.bar(x_bar,y_bar)
# plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6)) #.figsize(width,height)
plt.xticks(rotation=90,fontsize=10)
plt.bar(x_bar,y_bar,width=0.5)
plt.show()
```
